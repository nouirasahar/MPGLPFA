# This Notebook is for Data preprocessing

```python
import json
import random

# Load your original JSON data
with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

dialogues = []

for category, intents in data.items():
    for intent in intents:
        for pattern in intent["patterns"]:
            for response in intent["responses"]:
                # Create a short conversation with input and response
                dialogues.append({
                    "dialog": [
                        pattern.capitalize(),
                        response
                    ]
                })

# Shuffle and optionally sample a portion (e.g., to reduce size for training)
random.shuffle(dialogues)

# Save the formatted dataset
with open("dialoGPT_health_format.json", "w", encoding="utf-8") as f:
    for d in dialogues:
        f.write(json.dumps(d, ensure_ascii=False) + "\n,")
```

# PreProcessing

### Loading knowledge Base file with our custom Data

```python
fileStructure={
    "questions":[]
}
path="dialoGPT_health_format.json"
outputFilePath="structuredData.json"
with open(path,'r') as myfile:
    data=json.load(myfile)

for entry in data:
    items=entry.get("dialog")
    question=items[0]
    answer=items[1]
    questionObject={
        "question":question,
        "answer":answer
    }
    fileStructure["questions"].append(questionObject)

with open(outputFilePath,"w") as outputFile:
    outputFile.write(json.dumps(fileStructure,indent=4,ensure_ascii=False))

```
